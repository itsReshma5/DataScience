```python
import pandas as pd

# Replace with your actual file name after upload
df = pd.read_csv('/content/q1 dataset - Sheet1.csv')

# View first few rows
df.head()
```

```python
# a)Check missing values in key columns
print(df[['Location', 'Event Type', 'Casualties', 'Date']].isnull().sum())
```

```python
# Fill missing locations and event types with 'Unknown'
df['Location'].fillna('Unknown', inplace=True)
df['Event Type'].fillna('Unknown', inplace=True)

# Drop rows where 'casualties' or 'date' is missing (important numeric & time data)
df.dropna(subset=['Casualties', 'Date'], inplace=True)
print()
# Check missing values again after cleaning
print("\nMissing Values After Cleaning:")
print(df[['Location', 'Event Type', 'Casualties', 'Date']].isnull().sum())
```

```python
# b)Check for duplicates
print("Duplicates before:", df.duplicated().sum())

# Drop them
df = df.drop_duplicates()

print("Duplicates after:", df.duplicated().sum())
```

```python
# C)Convert and format dates
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')

print("\nSample Dates After Standardizing:")
print(df['Date'].head())
```

```python
# Print unique values of Event Type before cleaning
print("\nUnique Event Types Before Cleaning:")
print(df['Event Type'].unique())

# Convert 'Event Type', 'Sanction Type', and 'Social Media Sentiment' to lowercase and strip extra spaces
df['Event Type'] = df['Event Type'].str.strip().str.lower()
df['Sanction Type'] = df['Sanction Type'].str.strip().str.lower()
df['Social Media Sentiment'] = df['Social Media Sentiment'].str.strip().str.lower()

# Print unique values of Event Type after cleaning
print("\nUnique Event Types After Cleaning:")
print(df['Event Type'].unique())
```
